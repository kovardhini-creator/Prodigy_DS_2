```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a small sample Titanic-like dataset manually
data = {
    'sex': ['male', 'female', 'female', 'male', 'female', 'male', 'male', 'female'],
    'age': [22, 38, 26, 35, 27, 40, 19, 24],
    'survived': [0, 1, 1, 0, 1, 0, 0, 1]
}
df = pd.DataFrame(data)

# Plot a violin plot of Age vs Survival status
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='survived', y='age', palette='cool')
plt.title('Age Distribution by Survival Status (Violin Plot)', fontsize=14)
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.xlabel('Survival', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
